```python


# Target URL
url = "https://www.ambitionbox.com/list-of-companies?campaign=desktop_navpage%3D1"

# Headers to prevent 403 Forbidden errors by mimicking a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
}

try:
    # Send HTTP request to the page
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Check if the request was successful

    # Parse HTML content
    soup = BeautifulSoup(response.text, "html.parser")

    # List to store our scraped data dictionaries
    companies_list = []

    # AmbitionBox usually houses company listings inside specific card containers
    # (Note: Class names can change over time depending on website updates)
    company_cards = soup.find_all("div", class_="companyCardWrapper")

    # Fallback in case class structures changed: search for headers
    if not company_cards:
        company_cards = soup.find_all("div", attrs={"itemtype": "http://schema.org/LocalBusiness"}) 
        
    # If still empty, we parse by the structural h3 tags present on the page
    if not company_cards:
        print("Standard cards not found, parsing via alternative layout tags...")
        # Direct structural parsing based on current page source
        for h3 in soup.find_all("h3"):
            name = h3.text.strip()
            # Basic validation to skip footer/header h3 elements
            if name in ["Top Companies in INDIA", "Companies in India", "Get the AmbitionBox App"]:
                continue
                
            # Attempt to find siblings or parent info blocks
            parent = h3.find_parent("div")
            rating = "N/A"
            reviews = "N/A"
            industry = "N/A"
            
            if parent:
                rating_tag = parent.find("span", class_="rating") or parent.find("p")
                if rating_tag:
                    rating = rating_tag.text.strip()
                
                # Look for links containing review counts
                review_tag = parent.find("a", href=lambda x: x and "reviews" in x)
                if review_tag:
                    reviews = review_tag.text.strip()
            
            companies_list.append({
                "Company Name": name,
                "Rating": rating,
                "Reviews": reviews
            })
    else:
        # Loop through found card containers
        for card in company_cards:
            name = card.find("h3").text.strip() if card.find("h3") else "N/A"
            rating = card.find("span", class_="companyCardWrapper__companyRatingValue").text.strip() if card.find("span", class_="companyCardWrapper__companyRatingValue") else "N/A"
            reviews = card.find("span", class_="companyCardWrapper__ActionCount").text.strip() if card.find("span", class_="companyCardWrapper__ActionCount") else "N/A"
            
            # Extract basic info string (e.g., "IT Services & Consulting | Bengaluru...")
            info_tag = card.find("span", class_="companyCardWrapper__interconnectingDot")
            industry = info_tag.find_previous_sibling(text=True).strip() if info_tag else "N/A"

            companies_list.append({
                "Company Name": name,
                "Rating": rating,
                "Reviews": reviews,
                "Industry": industry
            })

    # Convert data into a structured Pandas DataFrame
    df = pd.DataFrame(companies_list)

    # Display data in the console
    print("\n--- Scraped Company Details ---")
    print(df)


except requests.exceptions.HTTPError as err:
    print(f"HTTP Error occurred: {err} (AmbitionBox may be blocking the request scale)")
except Exception as e:
    print(f"An error occurred: {e}")
```
