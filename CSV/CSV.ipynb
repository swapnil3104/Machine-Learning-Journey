CSV stands for Comma-Separated Values.

It stores tabular data in plain text.

Example students.csv:

id,name,age,city
1,Alice,20,Pune
2,Bob,22,Mumbai
3,Charlie,21,Delhi

First row = column names (header)
Remaining rows = data records
Values are separated by commas

Reading a CSV File
Basic Syntax

```python
import pandas as pd

df = pd.read_csv("student-data.csv")
print(df)
```

Viewing Data

```python

df.head() #first 5 rows
```

```python
df.tail() #last 5 rows
```

```python
df.info() #data types and non-null values
```

```python
df.describe() #statistical summary of numerical columns
```

```python
print(df.columns) #list of column names
```

Selecting Columns

```python
df["sex"] #access the "name" column
```

```python
df[["sex", "age"]] #access multiple columns
```

Accessing Rows
Using iloc (position)

```python
df.iloc[0, 1] #access first row, second column (age of first student)
```

Filtering Data
Students older than 20:

```python
df[df["age"] > 20] #filter rows where age is greater than 20
```

```python
df[df["sex"] == "M"] #filter rows where sex is "M"
```

Adding a New Column

```python
df["marks"] = [85, 90, 88]
```

Deleting Columns

```python
df.drop("Sex", axis=1)
```

Deleting Rows

Remove row index 1:

```python
df.drop(1)
```

```python
df.drop(1, inplace=True)
```

Dataset in chunks

```python
#loading a dataset in chunks
df_chunks = pd.read_csv("student-data.csv", chunksize=100)
```

```python
for chunk in df_chunks:
    print(chunk.shape)
```

usecolsSequence of Hashable or Callable, optional

```python
#usecolumns = ["name", "age"]

df = pd.read_csv("student-data.csv", usecols=["sex", "age"]) #load only "sex" and "age" columns
print(df.head())
```
